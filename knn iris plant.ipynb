```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import pandas as pd
```

```python
def knn(data,predict,k=3): 
    distances = []
    for group in data:
        for features in data[group]:
            distance = np.linalg.norm(np.array(features)-np.array(predict))
            distances.append([distance,group])
    votes = [i[1] for i in sorted(distances)[0:k]]
    vote_result = Counter(votes).most_common(1)[0][0]
    return vote_result
```

```python
df = pd.read_csv('Iris.csv') 
df.drop(['Id'],axis=1,inplace=True)
full_data = df.astype(float).values.tolist()
full_data
df
```

```python
x=[1,2,3,4,5]
test_size = 0.2
train_data = x[:-int(test_size*len(x))] 
test_data = x[-int(test_size*len(x)):]
test_data
```

```python
import random
random.shuffle(full_data) 
test_size = 0.2

train_data = full_data[:-int(test_size*len(full_data))]
test_data = full_data[-int(test_size*len(full_data)):]
len(train_data)
```

```python
len(full_data)
```

```python
train_set = {1:[] , 2:[] , 3:[] } 
test_set = {1:[] , 2:[] , 3:[] }
for i in train_data:
    train_set[i[-1]].append(i[:-1])
  
for i in test_data:
    test_set[i[-1]].append(i[:-1]) 
test_set
```

```python
knn(train_set,[6.1,3.7,6.1,1.9],7) # 1 = Iris-setosa
                                   #2 = Iris-versicolor
                                   #3 = Iris-virginica


```

```python
accuracy = [] 
z = [i for i in range(1,100,2)] 
for k in range(1,100,2):
    correct = 0
    total = len(test_data)
    for group in test_set:
        for data in test_set[group]: 
            vote = knn(train_set,data,k) 
            if group == vote:
                correct+=1
    accuracy.append(correct/total)
```

```python
z[accuracy.index(max(accuracy))]
```

```python
print(max(accuracy),z[accuracy.index(max(accuracy))]) 
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.plot(z,accuracy) 
plt.show() 
```
